# FOR THE DATABASE

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing data
data="https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv"
energy=pd.read_csv(data)
energy.head(5)
```

```python
#Cleaning the data
#To see all the information on the data
energy.info()
```

```python
energy.describe(include='all')
```

```python
#to see how many years are reported
energy.report_year.unique()
```

```python
#To find out if the fuel type codes
energy.fuel_type_code_pudl.unique()
```

```python
#grouping the data by their fuel type code and sum the for each year
fuel_data=pd.DataFrame(energy.groupby(['fuel_type_code_pudl','report_year'])[['fuel_qty_burned','fuel_mmbtu_per_unit','fuel_cost_per_unit_burned','fuel_cost_per_unit_delivered','fuel_cost_per_mmbtu']].sum())
fuel_data
```

```python
#Further simplifying the data
new_energy=energy[["fuel_qty_burned","fuel_mmbtu_per_unit","fuel_cost_per_unit_burned","fuel_cost_per_unit_delivered","fuel_cost_per_mmbtu"]]
new_energy
```

### HIGHEST AVERAGE FUEL COST PER UNIT DELIVERED

```python
#average fuel cost per unit delivered
fuel_avg=pd.DataFrame(energy.groupby(['report_year'])[['fuel_cost_per_unit_delivered']].mean())
fuel_avg
```

##### From the above, the highest average fuel cost per unit delivered is for the year 1997

### STANDARD VARIATION AND PERCENTILE

```python
#To find standard variation for fuel mmbtu per unit
standard_variation=new_energy["fuel_mmbtu_per_unit"].std()
standard_variation
```

```python
#To find percentile for fuel mmbtu per unit
percent_ile = np.percentile(new_energy["fuel_mmbtu_per_unit"], q=[75])
percent_ile
```

##### From the above, the standard variation to 2dp is 10.60 while the percentile to 2dp is 17.01

### SKEWNESS AND KURTOSIS

```python
#To find skewness for fuel quantity burned
skewness=new_energy["fuel_qty_burned"].skew(axis = 0, skipna = True)
skewness
```

```python
#to find kurtosis for fuel quantity burned
kurt=new_energy["fuel_qty_burned"].kurtosis()
kurt
```

##### From the above, the skewness and kurtosis to 2dp is 15.85 and 651.37 respectively

### MISSING VALUE PERCENTAGE

```python
#Checking for missing values
energy_missing=energy.isna().sum()
energy_missing
```

```python
len(energy)
```

```python
percentage_missing= (energy_missing/len(energy))*100
percentage_missing
```

##### From the above, we can see that the percentage of missing data is 0.61% in "fuel unit" which is a categorical column. The missing values can be filled using mode computation method

### 2nd and 3rd LOWEST CORRELATION OF FUEL COST PER UNIT BURNED

```python
#finding correlation
correlation = new_energy.corr()
correlation
```

##### From the above, we can see that the 2nd and 3rd lowest correlation is fuel cost per unit delivered and fuel qty burned

### PERCENTAGE CHANGE IN FUEL COST PER UNIT BURNED FOR THE YEAR 1994 and 1998

```python
#extracting values for year 1994
change=energy[energy['report_year'] == 1994]
change
```

```python
#extracting values for year 1998
change2=energy[energy['report_year'] == 1998]
change2
```

```python
#joining the two database together
fuel_change = pd.concat([change, change2])
fuel_change
```

```python
#grouping the fuel change database by report year
fuel=pd.DataFrame(fuel_change.groupby(['report_year'])[['fuel_cost_per_unit_burned']].sum())
fuel
```

```python
#finding the percent change

fuel.pct_change()
```

#### From the above, the percent change for fuel cost per unit burned for the year 1994 and 1998 is -46.7%
